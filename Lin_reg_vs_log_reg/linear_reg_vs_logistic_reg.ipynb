### Compare Linear & Logistic Regression
##### The idea in this demonstration is compare both model in a classification problem.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# we create random numbers and metrices with zeros and onex
x = np.array([i for i in range(10, 30)])
y = np.concatenate([np.ones(10), np.zeros(10)])
```

```python
# polot these numbers to represent values between zeros and ones
fig = plt.figure(figsize=(10,5))
xlabel = 'Age'
ylabel = 'Purchase'
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.grid(color='k', linestyle=':', linewidth=1)
_ = plt.plot(x, y, 'xb')
```

```python
# fit the best linear model line
lin_regression = LinearRegression()
_ = lin_regression.fit(pd.DataFrame(x), y)

lin_y_pred_1 = lin_regression.predict(pd.DataFrame(x))

line_point_5 = x * 0 + 0.5
    
fig = plt.figure(figsize=(10,5))
xlabel = 'Age'
ylabel = 'Purchase'
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.grid(color='k', linestyle=':', linewidth=1)
plt.plot(x, y, 'xb')
plt.plot(x, lin_y_pred_1, '-r')
_ = plt.plot(x, line_point_5,'-g')
```
